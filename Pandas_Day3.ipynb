```python
# join matches the indexes matching
# merge check for the column values (even though column name is same , we can able to do merge function)
```

```python
import pandas as pd
```

```python
d1 = {'key1':[1,2,3,4,5,"ayesha"],
      'key2' : [7,8,9,10,11,12],
      'key3':[13,14,15,16,17,18],
      'key4':[19,20,21,22,23,24]
}
```

```python
df1=pd.DataFrame(d1,index=['a','b','c','d','e','f'])
```

```python
d2 = {'key1':[1,2,3,4,5,"ayesha"],
      'key2' : [7,8,"mafrah",10,11,12],
      'key3':[13,14,15,16,17,18],
      'key4':[19,20,21,"python","pandas",24]
}
```

```python
df2=pd.DataFrame(d2)
```

```python
df1
```

```python
df2
```

```python
pd.merge(df1,df2)
```

```python
pd.merge(df1,df2,how="inner")
```

```python
pd.merge(df1,df2,how="inner",on="key1")
```

```python
pd.merge(df1,df2,how="inner",on="key4")
```

```python
pd.merge(df1,df2,how="left",on="key4")
```

```python
pd.merge(df1,df2,how="right",on="key4")
```

```python
pd.merge(df1,df2,how="outer",on="key4")
```

```python
pd.merge(df1,df2,how= 'outer',on= ['key4','key1'])
```
